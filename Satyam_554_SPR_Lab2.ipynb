# Q1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frequency = 5
samplingrate = 1000
duration = 5
```

```python
t = np.arange(0, duration, 1/samplingrate)
```

```python
amplitude = 1
sinusoidalsignal = amplitude * np.sin(2 * np.pi * frequency * t)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(t, sinusoidalsignal)
plt.title(f'Sine Wave: {frequency}Hz, Sampled at {samplingrate}Hz')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
dtft = np.fft.fft(sinusoidalsignal)
n = len(sinusoidalsignal)
```

```python
n
```

```python
frequencies = np.fft.fftfreq(n, 1/samplingrate)
```

```python
frequencies
```

```python
plt.figure(figsize=(10, 5))

pos_frequencies = frequencies[:n//2]
mag_spectrum = np.abs(dtft)[:n//2]

plt.plot(pos_frequencies, mag_spectrum)
plt.title('Discrete-Time Fourier Transform (DTFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
dft = np.fft.fft(sinusoidalsignal)
```

```python
plt.figure(figsize=(10, 5))

mag_spectrum = np.abs(dft)[:n//2]

plt.stem(pos_frequencies, mag_spectrum, use_line_collection=True)
plt.title('Discrete Fourier Transform (DFT) - Discrete Frequency Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

# Q2

```python
SR1 = 1000
duration1 = 1
```

```python
t = np.arange(0, duration1, 1/SR1)
```

```python
frequency1 = 5
amplitude1 = 1

frequency2 = 15
amplitude2 = 0.5

frequency3 = 30
amplitude3 = 0.8
```

```python
sin_wave1 = amplitude1 * np.sin(2 * np.pi * frequency1 * t)
sin_wave2 = amplitude2 * np.sin(2 * np.pi * frequency2 * t)
sin_wave3 = amplitude3 * np.sin(2 * np.pi * frequency3 * t)
```

```python
compositesignal = sin_wave1 + sin_wave2 + sin_wave3
```

```python
plt.figure(figsize=(10, 5))
plt.plot(t, compositesignal)
plt.title('Composite Signal: Sum of Multiple Sine Waves')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
dtft = np.fft.fft(compositesignal)
n = len(compositesignal)
```

```python
frequencies1 = np.fft.fftfreq(n, 1/SR1)
```

```python
plt.figure(figsize=(10, 5))

pos_frequencies1 = frequencies1[:n//2]
mag_spectrum1 = np.abs(dtft)[:n//2]

plt.plot(pos_frequencies1, mag_spectrum1)
plt.title('Discrete-Time Fourier Transform (DTFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
dft = np.fft.fft(compositesignal)
plt.figure(figsize=(10, 5))

mag_spectrum2 = np.abs(dft)[:n//2]

plt.stem(pos_frequencies1, mag_spectrum2, use_line_collection=True)
plt.title('Discrete Fourier Transform (DFT) - Discrete Frequency Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
amplitude = 1
decay_constant = 2
frequency = 5
sampling_rate = 1000
duration = 2
```

```python
t = np.arange(0, duration, 1/sampling_rate)
```

```python
decaying_signal = amplitude * np.exp(-decay_constant * t) * np.sin(2 * np.pi * frequency * t)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(t, decaying_signal)
plt.title('Exponentially Decaying Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
dtft = np.fft.fft(decaying_signal)
n = len(decaying_signal)
```

```python
frequencies2 = np.fft.fftfreq(n, 1/sampling_rate)
```

```python
plt.figure(figsize=(10, 5))

pos_frequencies2 = frequencies2[:n//2]
mag_spectrum2 = np.abs(dtft)[:n//2]

plt.plot(pos_frequencies2, mag_spectrum2)
plt.title('Discrete-Time Fourier Transform (DTFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
dft = np.fft.fft(decaying_signal)
plt.figure(figsize=(10, 5))

mag_spectrum2 = np.abs(dft)[:n//2]

plt.stem(pos_frequencies2, mag_spectrum2, use_line_collection=True)
plt.title('Discrete Fourier Transform (DFT) - Discrete Frequency Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```
